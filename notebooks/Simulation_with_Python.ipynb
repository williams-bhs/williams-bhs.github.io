# Stick Break
Given a stick broken randomly into three pieces, what is the probability that the pieces can be formed into a triangle?

```python
# Break stick at two places
break1 = random()
break2 = random()

# Determine lengths of pieces based on breaks
piece1 = min( break1, break2 )
piece2 = max( break1, break2 ) - piece1
piece3 = 1 - max( break1, break2 )

# Print lengths
print( "Piece 1 = ", piece1 )
print( "Piece 2 = ", piece2 )
print( "Piece 3 = ", piece3 )
```

## Testing for a Triangle
Now that we have the three piece lengths, let's determine if they can make a triangle or not. For three pieces $a$, $b$, and $c$, a triangle can be formed if $a+b>c$, $a+c>b$, and $b+c>a$.

```python
# Test the triangle
if piece1 + piece2 > piece3 and piece1 + piece3 > piece2 and piece2 + piece3 > piece1:
  print( "Triangle!" )
else:
  print( "No triangle :(" )
```

# Repeating a Simulation with a Loop
A loop is a mechanism for repeating a sequence of commands.

```python
# In this loop, x starts at 0 and the loop repeats until x reaches 10 (and stops)
for x in range(10):
  print( x )
```

Instead of printing $x$, lets flip a coin. The code that is contained within the loop is all indented relative to the `for` statement.

```python
for x in range(10):
  if random() < 0.5:
    print("Heads")
  else:
    print("Tails")
```

The following loop contains several commands that are all indented relative to the `for` statement.

```python
N = 1000
triangles = 0

# Repeat the stick breaking N times
for x in range(N):

  # Break stick at two places
  break1 = random()
  break2 = random()

  # Determine lengths of pieces based on breaks
  piece1 = min( break1, break2 )
  piece2 = max( break1, break2 ) - piece1
  piece3 = 1 - max( break1, break2 )

  # Test for a triangle
  if piece1 + piece2 > piece3 and piece1 + piece3 > piece2 and piece2 + piece3 > piece1:
    triangles += 1

# Print the proportion of triangles
print( triangles / N )
```
