# Penguins Dataset Overview

The data below was collected on **three species of penguin**:

penguins.png

Source: https://allisonhorst.github.io/palmerpenguins/

The dataset contains values for:
1. Species
2. Island
3. Beak Length (mm)
4. Beak Depth (mm)
5. Flipper Length (mm)
6. Body Mass (g)
7. Sex (Male or Female)

We will work only with the four numerical columns. Importantly,
- The first 152 rows are **aedeli**.
- The next 68 rows are **chinstrap**.
- The final 124 rows are **gentoo**.



````python
# This cell imports libraries and downloads the dataset.
import numpy as np
import matplotlib.pyplot as plt

# Download penguin data
import kagglehub
path = kagglehub.dataset_download("amulyas/penguin-size-dataset")

# Read the CSV data into numpy, only capturing the four numerical columns
data = np.genfromtxt(
  path+'/penguins_size.csv',
  delimiter=',',
  skip_header=1,
  usecols=(2, 3, 4, 5)
)
````
